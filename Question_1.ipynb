```python
!pip install  wandb
import wandb
from wandb.keras import WandbCallback
from keras.datasets import fashion_mnist
wandb.login()
wandb.init(project="cs6910_assignment1",entity="cs21s006_cs21s043")

(train_X,train_Y),(test_X,test_Y)=fashion_mnist.load_data()
classes=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
img=[]
lbl=[]

for i in range(len(test_X)):
  if classes[test_Y[i]] not in lbl:
    img.append(test_X[i])
    lbl.append(classes[test_Y[i]])

  if len(lbl)==10:
    break

wandb.log({"one image from each class(For test images)": [wandb.Image(i, caption=l) for i,l in zip(img,lbl)]})

```
